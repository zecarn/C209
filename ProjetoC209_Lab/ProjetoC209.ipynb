- Importar bibilotecas

```python
import moviepy.editor as mpe
from moviepy.video.fx.mirror_x import mirror_x
from moviepy.audio.fx.all import volumex
```

- Carregar o vídeo a ser utilizado

```python
video = mpe.VideoFileClip("video_projeto.mp4")

print(f"Tamanho: {video.size}")
print(f"FPS: {video.fps}")
print(f"Duração: {video.duration} segundos")
print(f"Numero de frames: {video.reader.nframes}")
```

- Inverter a imagem no eixo a cada 20 segundos

```python
segments = []
duration = video.duration
for start in range(0, int(duration), 20):
    # dividir em intervalos de 20 segundos
    segment = video.subclip(start, start + 20)

    if (start // 20) % 2 == 1:
        segment = mirror_x(segment)  # inverter no eixo horizontal
        
    segments.append(segment)

video = mpe.concatenate_videoclips(segments)
```

- Reduzir o som gradativamente a cada 30 segundos

```python
audio = video.audio
audio_duration = audio.duration
volume_reduction = []

for start in range(0, int(audio_duration), 30):
    # dividir em intervalos de 30 segundos
    subclip = audio.subclip(start, start + 30)
    
    # ajustar volume gradualmente no último intervalo
    if start + 30 >= audio_duration - 10:
        # audio sem som nos ultimos 10 segundos
        subclip = volumex(subclip, 0)
    else:
        # reduzir o volume a cada subclip
        subclip = volumex(subclip, 1 - round(start / audio_duration, 2))  
        print(1 - round(start / audio_duration, 2)) # redução proporcional ao progresso
    
    volume_reduction.append(subclip)

# recombinar o áudio ajustado
audio = mpe.concatenate_audioclips(volume_reduction)
video = video.set_audio(audio)
```

- Fazer um corte entre o segundo 60 até o 80


```python
part1 = video.subclip(0, 60)
cut_part = video.subclip(60, 80)
part2 = video.subclip(80, duration)
```

- Inserir os 20 segundos cortados acima no fim do vídeo


```python
final_video = mpe.concatenate_videoclips([part1, part2, cut_part])
```

- Salvar o resultado final do vídeo

```python

final_video.write_videofile("projeto_final.mp4", codec="libx264", audio_codec="aac")

print("Vídeo final salvo como 'projeto_final.mp4'")
```
